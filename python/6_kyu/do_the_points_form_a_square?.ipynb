You are given a list of four points (x, y). Each point is a tuple

Your task is to write a function which takes an array of points and returns whether they form a square.

If the array doesn't contain 4 points, return False.

All points coordinates are integers.

Squares can be rotated using a random degree.

Examples
```
((1,1), (3,3), (1,3), (3,1)) -> True
((0, 0), (0,2), (2,0), (2,1)) -> False
([]) -> False
```
Constraints:
```
1 <= x, y <= 1000
```
Note:

Performance does not enforced in this kata.

How to find diagonal of square by side
```python
diagonal = ((side ** 2)*2) ** 0.5
```
how to find distance between two points

```python
distance = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5
```

````python
# find distance between any two points
def dist(points:tuple) -> float:
    '''
    param:points Tuple with two points of square
    '''
    # define x and y for every point
    x1 = points[0][0]
    y1 = points[0][1]
    x2 = points[1][0]
    y2 = points[1][1]
    
    # calculate distance
    distance = ((x2 - x1)**2 + (y2 - y1)**2) ** 0.5
    return distance


# check for square
def is_square(points):
    
    if len(points) != 4:
        return False
    
    a = points[0]
    b = points[1]
    c = points[2]
    d = points[3]
    

    distances = set()
    
    points_combinations = [(a, b), (a, d), (b,c), (c,d), (a, c), (b, d)]
    
    # keep distance betwee points
    for i in points_combinations:
        distances.add(dist(i))
        
    # distance between all sides should be equal. And diagonals should be equal
    # so we need only distance between side points and between diagonal points
    # and set wil keep two values for square because others should be the same
    if len(distances) == 2:
        # smalles value is side
        side = min(distances)
        # biggest value is diagonal
        diagonal = max(distances)
        # if diagonal is diagonal of side
        if round(((side ** 2)*2) ** 0.5, 2) == round(diagonal, 2):
            return True
        else:
            return False
        
    else:
        return False
    
print(is_square([(891, 891), (316, 892), (890, 316), (315, 317)]))
````
